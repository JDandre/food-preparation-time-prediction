```python
import pandas as pd

train = pd.read_csv("train_clean.csv")
```

```python
food_quant_labels = []

for col in train.columns:
    #bigger for guid length
    if("quant_" in col and len(col) > 30):
        food_quant_labels.append(col)
        
def normalize_food_quants(arr):
    for i in range(0, len(arr)):
        arr[i] /= 3 # divide by max val
    return arr
```

```python
food_type_quant_labels = []
for col in train.columns:
    # smaller than guid length
    if("quant_" in col and len(col) < len("quant_f7f996af-8133-408234234")):
        print("Max value of", col, ": ", train[col].max())
        food_type_quant_labels.append(col)

def normalize_food_type_quants(arr):
    # No normalization for now (binary maybe?)
    return arr
```

```python
import math

times_labels = ["day", "hour", "minute"]

def normalize_times(arr):
    new_arr = []
    new_arr.append(math.sin(arr[0] / 7))
    new_arr.append(math.sin(arr[1] / 24))
    new_arr.append(math.sin(arr[2] / 60))
    
    new_arr.append(math.cos(arr[0] / 7))
    new_arr.append(math.cos(arr[1] / 24))
    new_arr.append(math.cos(arr[2] / 60))
    
    return new_arr
```

```python
# Predicting value
yhat_label = "food_prep_time_minutes"
```

```python
print("Number of inputs (predictors): ")
print("All food quantities: ", len(food_quant_labels))
print("Food types quantities: ", len(food_type_quant_labels))
print("Time details:", len(times_labels))
print("Total: ", len(food_quant_labels) + len(food_type_quant_labels) + len(times_labels))
```

```python
def preprocess_data(row, isTrain = True):
    #### Inputs
    inputs = []
    
    # Add time to inputs array
    time_inputs = []
    for times_label in times_labels:
        time_inputs.append(row[times_label])
    time_inputs = normalize_times(time_inputs)
    inputs += time_inputs
    
    # Add food quantities to inputs array
    food_quant_inputs = []
    for food_quant_label in food_quant_labels:
        food_quant_inputs.append(row[food_quant_label])
    food_quant_inputs = normalize_food_quants(food_quant_inputs)
    inputs += food_quant_inputs
    
    # Add quantities for different food types to inputs array
    food_type_quant_inputs = []
    for food_type_quant_label in food_type_quant_labels:
        food_type_quant_inputs.append(row[food_type_quant_label])
    food_type_quant_inputs = normalize_food_type_quants(food_type_quant_inputs)
    inputs += food_type_quant_inputs
    
    if(isTrain):
        ### Outputs
        outputs = []
        
        # Add actual preperation time to outputs array
        outputs.append(row[yhat_label])
    
        return inputs, outputs
    else:
        return inputs
```

```python
import numpy as np

x_all = []
y_all = []

for index, row in train.iterrows():
    inputs, outputs = preprocess_data(row, True)
    x_all.append(np.array(inputs))
    y_all.append(np.array(outputs))
    
x_all = np.array(x_all)
y_all = np.array(y_all)
```

```python
print("Input sample #1:", x_all[1])
print("Output sample #1:", y_all[1])
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Split the data to 80% train data, 20% test data
x_train, x_valid, y_train, y_valid = train_test_split(x_all, y_all,
                                                      test_size = 0.2, shuffle = True, random_state=42)

print("Train data size:", len(x_train), "Test data size:", len(x_valid))
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout

input_size = len(food_quant_labels) + len(times_labels) * 2 + len(food_type_quant_labels)

print("Input size: ", input_size)

model = Sequential()
model.add(Dense(256, input_dim=109, activation="relu"))
model.add(BatchNormalization())
model.add(Dense(192, activation="relu"))
model.add(BatchNormalization())
model.add(Dense(128, activation="relu"))
model.add(BatchNormalization())
model.add(Dense(96, activation="relu"))
model.add(BatchNormalization())
model.add(Dense(1, activation="relu")) # output layer

model.summary()
```

```python
from tensorflow.keras import optimizers
import tensorflow as tf

model.compile("adam", #optimizers.SGD(learning_rate = 0.01, momentum = 0.8, nesterov=True),
              loss='mean_squared_error',
              metrics=['mae', 'mse'])
```

```python
import numpy as np

input_size = len(x_train[0])

# first hidden layer size 1024
epochs = 100
model.fit(np.array(x_train), np.array(y_train),
          batch_size=100, # Number of samples to calculate before each weight update
          validation_data = (np.array(x_valid), np.array(y_valid)), 
          epochs=epochs)
```

```python
y_preds = model.predict(x_train)
y_preds = y_preds.reshape(y_preds.shape[0])

y_vals = np.array(y_train)
y_vals = y_vals.reshape(y_vals.shape[0]) # flatten the values

print_count = 25

# Print out first 25 validation predicted values and actual values as a sample
for i in range(0, print_count):
    print("Predicted: ", y_preds[i], ", Actual: ", y_vals[i])
```

```python
import math
import scipy
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

# https://stackoverflow.com/a/13918150
def rsquared(x, y):
    """ Return R^2 where x and y are array-like. """

    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
    return r_value**2

print("r2 =", rsquared(y_vals, y_preds), "> 0.6100377")
print("mae =", mean_absolute_error(y_vals, y_preds), "< 5.448831")
print("root mse =", math.sqrt(mean_squared_error(y_vals, y_preds)), "< 6.978090")
```
